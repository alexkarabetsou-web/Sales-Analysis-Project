```python
import pandas as pd

url = "https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_April_2019.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df = df.dropna()

df.info()
```

```python
df.head(20)
```

```python
df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"])
df["Price Each"] = pd.to_numeric(df["Price Each"])

df.info()
```

```python
df = df[df["Quantity Ordered"] != "Quantity Ordered"]

df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"])
df["Price Each"] = pd.to_numeric(df["Price Each"])

df.info()
```

```python
df["Sales"] = df["Quantity Ordered"] * df["Price Each"]

df.head()
```

```python
df["Sales"].sum()
```

```python
df.groupby("Product")["Sales"].sum().sort_values(ascending=False)
```

```python
product_sales = df.groupby("Product")["Sales"].sum().sort_values()

product_sales.plot(kind="barh", figsize=(10,6))

plt.title("Revenue by Product")
plt.xlabel("Revenue ($)")
plt.ylabel("Product")

plt.show()
```

```python
df.groupby("Product")["Quantity Ordered"].sum().sort_values(ascending=False)
```

```python
product_quantity = df.groupby("Product")["Quantity Ordered"].sum().sort_values()

product_quantity.plot(kind="barh", figsize=(10,6))

plt.title("Units Sold by Product")
plt.xlabel("Units Sold")
plt.ylabel("Product")

plt.show()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df.info()
```

```python
df["Month"] = df["Order Date"].dt.month

df.head()
```

```python
df["Hour"] = df["Order Date"].dt.hour

df.head()
```

```python
hours = df.groupby("Hour").size()

hours
```

```python
plt.figure(figsize=(10,5))

plt.plot(hours.index, hours.values)

plt.title("Orders by Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Orders")

plt.grid()

plt.show()
```

```python
def get_city(address):
    return address.split(",")[1]

df["City"] = df["Purchase Address"].apply(get_city)

df.head()
```

```python
city_sales = df.groupby("City")["Sales"].sum().sort_values(ascending=False)

city_sales
```

```python
plt.figure(figsize=(10,5))

city_sales.plot(kind="bar")

plt.title("Revenue by City")
plt.xlabel("City")
plt.ylabel("Revenue ($)")

plt.xticks(rotation=45)

plt.show()
```

```python
df_dupes = df[df["Order ID"].duplicated(keep=False)]

df_dupes.head()
```

```python
df_dupes["Grouped"] = df_dupes.groupby("Order ID")["Product"].transform(lambda x: ",".join(x))

df_dupes.head()
```

```python
df_dupes = df_dupes[['Order ID', 'Grouped']].drop_duplicates()

df_dupes.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df_dupes["Grouped"]:
    products = row.split(",")

    count.update(combinations(products, 2))

count.most_common(10)
```
